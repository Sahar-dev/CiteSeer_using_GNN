```python
def train():
      model.train()
      optimizer.zero_grad()  # Clear gradients.
      out = model(data.x, data.edge_index)  # Perform a single forward pass.
      loss = criterion(out[data.train_mask], data.y[data.train_mask])  # Compute the loss solely based on the training nodes.
      loss.backward()  # Derive gradients.
      optimizer.step()  # Update parameters based on gradients.
      return loss
```

```python
def test():
      model.eval()
      out = model(data.x, data.edge_index)
      pred = out.argmax(dim=1)  # Use the class with highest probability.
      test_correct = pred[data.test_mask] == data.y[data.test_mask]  # Check against ground-truth labels.
      test_acc = int(test_correct.sum()) / int(data.test_mask.sum())  # Derive ratio of correct predictions.
      return test_acc
```

## GCN V2

```python
from torch_geometric.nn import GATConv

class GAT(torch.nn.Module):
    def __init__(self, hidden_channels, heads):
        super().__init__()
        torch.manual_seed(1234567)
        self.conv1 = GATConv(dataset.num_features, hidden_channels, heads=heads, dropout=0.6)
        self.conv2 = GATConv(hidden_channels * heads, dataset.num_classes, heads=1, concat=False, dropout=0.6)

    def forward(self, x, edge_index):
        x = F.dropout(x, p=0.6, training=self.training)
        x = self.conv1(x, edge_index)
        x = F.elu(x)
        x = F.dropout(x, p=0.6, training=self.training)
        x = self.conv2(x, edge_index)
        return x

model = GAT(hidden_channels=8, heads=8)
print(model)

optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

def train():
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss

def test(mask):
    model.eval()
    out = model(data.x, data.edge_index)
    pred = out.argmax(dim=1)
    correct = pred[mask] == data.y[mask]
    acc = int(correct.sum()) / int(mask.sum())
    return acc

for epoch in range(1, 201):
    loss = train()
    val_acc = test(data.val_mask)
    test_acc = test(data.test_mask)
    print(f'Epoch: {epoch:03d}, Loss: {loss:.4f}, Val: {val_acc:.4f}, Test: {test_acc:.4f}')
```

```python
model.eval()

out = model(data.x, data.edge_index)
visualize(out, color=data.y)
```

```python
test_acc = test()
print(f'Test Accuracy: {test_acc:.4f}')
```

## Random walk

```python
def random_walk(graph, start_node, walk_length):
    walk = [start_node]
    for _ in range(walk_length - 1):
        neighbors = list(graph.neighbors(walk[-1]))
        if neighbors:
            walk.append(random.choice(neighbors))
        else:
            break
    return walk
```

```python
# Convert PyG graph to NetworkX graph
G = nx.Graph()
G.add_edges_from(data.edge_index.t().numpy())
```

```python
# Filter out invalid nodes from random walks
valid_nodes = set(G.nodes)
random_walks = [random_walk(G, node, walk_length=10) for node in G.nodes if node in valid_nodes]
```

```python
# Train Node2Vec only on valid nodes
model = Node2Vec(G, dimensions=64, walk_length=10, num_walks=100, workers=4)
model.fit(window=5, min_count=1)
```

```python
from gensim.models import Word2Vec
# Generate random walks using Node2Vec
walks = Node2Vec(G, dimensions=64, walk_length=30, num_walks=200, workers=4).walks

# Convert walks to list of strings
walks = [list(map(str, walk)) for walk in walks]

# Train Word2Vec model on the walks
model = Word2Vec(walks, vector_size=64, window=10, min_count=1, sg=1, workers=4, epochs=5)

# Get node embeddings
node_embeddings = {str(node): model.wv[str(node)] for node in G.nodes}


# Convert embeddings to PyTorch tensor
node_embeddings_tensor = torch.tensor([node_embeddings[str(node)] for node in G.nodes], dtype=torch.float)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import normalized_mutual_info_score
labels = data.y.numpy()
# Convert embeddings to numpy array
embeddings_numpy = node_embeddings_tensor.numpy()

# Specify the number of clusters (replace with your desired number)
num_clusters = 6

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_assignments = kmeans.fit_predict(embeddings_numpy)
```

```python
# Use the minimum length to ensure both arrays have the same length
min_length = min(len(labels), len(cluster_assignments))

# Truncate both arrays to the minimum length
labels = labels[:min_length]
cluster_assignments = cluster_assignments[:min_length]
```

```python
# Evaluate clustering using normalized mutual information (NMI)
nmi = normalized_mutual_info_score(labels, cluster_assignments)
print(f'Normalized Mutual Information: {nmi:.4f}')
```

The NMI score ranges from 0 to 1, where:

0 indicates no mutual information (no similarity between the clusterings).

1 indicates perfect agreement between the clusterings.

The NMI score of 0.0039 is very close to 0, suggesting a low level of similarity between the true labels and the cluster assignments. This could mean that the clustering algorithm did not perform well in capturing the underlying structure present in the true labels.

```python
import networkx as nx
import torch
from gensim.models import Word2Vec
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Convert PyG graph to NetworkX graph
G = nx.Graph()
G.add_edges_from(data.edge_index.t().numpy())

# Train Node2Vec only on valid nodes
model_node2vec = Node2Vec(G, dimensions=64, walk_length=10, num_walks=100, workers=4)
model_node2vec.fit(window=5, min_count=1)

# Generate random walks using Node2Vec
walks = model_node2vec.walks

# Convert walks to list of strings
walks = [list(map(str, walk)) for walk in walks]

# Train Word2Vec model on the walks
model_word2vec = Word2Vec(walks, vector_size=64, window=10, min_count=1, sg=1, workers=4, epochs=5)

# Get node embeddings
node_embeddings = {str(node): model_word2vec.wv[str(node)] for node in G.nodes}

# Convert embeddings to PyTorch tensor
node_embeddings_tensor = torch.tensor([node_embeddings[str(node)] for node in G.nodes], dtype=torch.float)
```

```python
# Assuming you have node labels (ground truth) for all nodes
labels = data.y.numpy()

# Convert labels to PyTorch tensor
labels_tensor = torch.tensor(labels, dtype=torch.long)

# Ensure that the number of nodes and labels match
node_indices = [int(node) for node in G.nodes]
labels_tensor = labels_tensor[node_indices]

# Split the data into training and testing sets
train_embeddings, test_embeddings, train_labels, test_labels = train_test_split(
    node_embeddings_tensor.numpy(), labels_tensor.numpy(), test_size=0.2, random_state=42
)

# Train a random forest classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(train_embeddings, train_labels)

# Predict on the test set
predictions = classifier.predict(test_embeddings)

# Evaluate the classifier
accuracy = accuracy_score(test_labels, predictions)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
